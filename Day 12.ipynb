```python
#data processing
f=open("Day 12.txt", "r")
batch = f.read().split('\n') #Split by new lines
batch = [(x[0], int(x[1:])) for x in batch]
print(batch)
```

## Part 1

```python
def nsew(d, v):
    delta = (0, 0)
    if d == 'N': delta = (0,-v)
    if d == 'S': delta = (0,v)
    if d == 'E': delta = (-v,0)
    if d == 'W': delta = (v,0)
    
    return delta
```

```python
def changeDir(d, v, currDir):
    #print("change direction!")
    compass = ['N', 'E', 'S', 'W'] 

    ind = int(v / 90)
    mult = -1 if d == 'L' else 1

    return compass[(compass.index(currDir)+ind*mult) % len(compass)]
```

```python
def nav(dt, currDir, coord):
    
    (d, v) = dt
    delta = (0, 0)

    #print("Coord", coord, "Dir", currDir)
    #print("Nav",d, v)
    
    if d in ['N','E','S','W']: delta = nsew(d, v)
    if d == 'F': delta = nsew(currDir, v)
    if d in ['L','R']: currDir = changeDir(d, v, currDir)
        
    #print("Delta", delta,"\n")
    return (tuple(sum(x) for x in zip(coord, delta)), currDir)
```

```python
#Part 1 Run
coord = (0, 0)
currDir = 'E'

for dt in batch:
    coord, currDir = nav(dt, currDir, coord)

print("Final Coordinates",coord)
print("Manhattan Distance", abs(coord[0]) + abs(coord[1]))
```

## Part 2

```python
def changeDir2(d, v, waypoint):

    #how many times to rotate clockwise
    num = (int(v / 90) * (-1 if d == 'L' else 1)) % 4
    
    for i in range(num):
        waypoint = tuple([waypoint[1], -waypoint[0]])
        
    return waypoint
```

```python
def nav2(dt, coord, waypoint):

    (d, v) = dt
    delta = (0, 0)
    
    #print("Coord", coord)
    #print("Nav",d, v)
    
    if d in ['N','E','S','W']: waypoint = tuple(sum(x) for x in zip(nsew(d, v),waypoint))
    if d == 'F': delta = tuple(v*x for x in waypoint)
    if d in ['L','R']: waypoint = changeDir2(d, v, waypoint)

    #print("Delta", delta)
    #print("Waypoint", waypoint,"\n")
    return (tuple(sum(x) for x in zip(coord, delta)), waypoint)
```

```python
#Part 2 Run
coord = (0, 0)
waypoint = (-10, -1)

for dt in batch:
    coord, waypoint = nav2(dt, coord, waypoint)

print("Final Coordinates",coord)
print("Manhattan Distance", abs(coord[0]) + abs(coord[1]))
```
