```python
f=open("Day 14.txt", "r")
batch = f.read().split('\n') #Split by new lines

batch = [x.split(" = ") for x in batch]

print(batch)
```

```python
#Part 1
def binaryConverter(num):
    binary = "{0:b}".format(num)
    arr = [0] * 36
    arr[-len(binary):] = binary
    return [int(x) for x in arr]

def decimalConverter(b):
    return sum([x * 2**(len(b)-i-1) for i, x in enumerate(b)])

def applyMask(binary, mask):

    for i in range(len(mask)):
        if mask[i] != 'X': binary[i] = int(mask[i])
    
    return binary
```

```python
mask = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
mem_dict = dict()

for item in batch:
    
    #update mask
    if item[0] == 'mask':
        mask = item[1]
        continue
    
    b_masked = applyMask(binaryConverter(int(item[1])), mask)
    key = item[0].split("[")[1][:-1]
    mem_dict.update( {key : decimalConverter(b_masked)} )

print(sum([value for key, value in mem_dict.items()]))
#print(mem_dict)
```

```python
def applyMask2(binary, mask):
    
    for i in range(len(mask)):
        if mask[i] == 'X': binary[i] = 'X'
        if mask[i] == '1': binary[i] = 1
    
    return binary

def floatingMemSet(binary, mem_set):
    
    if 'X' not in binary: #no more Xs
        mem_set.add(decimalConverter(binary))
        return mem_set
    else:
        i = binary.index('X')
        v1 = binary[:i] + [1] + binary[i+1:]
        v0 = binary[:i] + [0] + binary[i+1:]
        return floatingMemSet(v1, mem_set).union(floatingMemSet(v0, mem_set))
```

```python
mask = '000000000000000000000000000000000000'
mem_dict = dict()

for item in batch:
    
    #update mask
    if item[0] == 'mask':
        mask = item[1]
        continue
    
    #get list of memory slots
    mem = int(item[0].replace('mem', '').replace('[','').replace(']',''))
    mem_masked = applyMask2(binaryConverter(mem), mask) 
    mem_set = floatingMemSet(mem_masked, set())
    #print(mem_set)
    
    #add to memory dictionary
    for key in mem_set:
        mem_dict.update( {key : int(item[1])} )

print(sum([value for key, value in mem_dict.items()]))
print(mem_dict)
```
